```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df
```

```python
X = df.drop('salary_more_then_100k', axis = True)
y = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job =  LabelEncoder()
le_degree =  LabelEncoder()
```

```python
X['company_n'] = le_company.fit_transform(X['company'])
X['job_n'] = le_company.fit_transform(X['job'])
X['degree_n'] = le_company.fit_transform(X['degree'])
X
```

```python
X_n = X.drop(['company','job','degree'], axis = True)
X_n
```

```python
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X_n, y, train_size = 0.8, random_state = 42) 
dt.fit(X_train,y_train)
dt.score(X_test,y_test)
```
