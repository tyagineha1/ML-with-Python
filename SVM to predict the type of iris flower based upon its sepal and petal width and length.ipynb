```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data)
print(iris.data.shape)
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target == 0]
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
import seaborn as sns
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
df2.head()
```

```python
sns.jointplot(x="sepal length (cm)",y="sepal width (cm)",data=df0,kind="scatter", color = 'red')
sns.jointplot(x="sepal length (cm)",y="sepal width (cm)",data=df1,kind="scatter", color = 'blue')
```

```python
X = df.drop(['target','flower_name'], axis = True)
X
y = df.target
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42) 
len(X_train)
len(X_test)
```

```python
from sklearn.svm import SVC
svc = SVC(gamma=1)
svc.fit(X_train,y_train)
svc.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
svc = SVC(C=100)
svc.fit(X_train,y_train)
svc.score(X_test,y_test)
```
